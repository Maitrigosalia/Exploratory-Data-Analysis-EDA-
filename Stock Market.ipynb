```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("C:\\Users\\INDIA\\Downloads\\National_Stock_Exchange_of_India_Ltd.csv")
df

df=df.head(5)
x=df['Symbol']
y=df['Open']
y1=df['High']
y2=df['Low']
y4=df['LTP']

#plt.figure(figsize=(8,6))
plt.subplot(2,2,4)
plt.scatter(x,y,color="green",label="open price")
#plt.xticks(rotation=90)
plt.grid(True)
plt.legend()
#plt.figure(figsize=(8,6))
plt.subplot(2,2,1)
plt.scatter(x,y1,color="red",label="high price")
#plt.xticks(rotation=90)
plt.grid(True)
plt.legend()

plt.subplot(2,2,2)
plt.scatter(x,y2,color="red",label="low price")
#plt.xticks(rotation=90)
plt.grid(True)
plt.legend()

plt.subplot(2,2,3)
plt.scatter(x,y4,color="red",label="ltp price")
#plt.xticks(rotation=90)
plt.grid(True)
plt.legend()
plt.show()
```
