```python
#Q1 Survival Rate Analysis
#•	What is the overall survival rate of passengers?

a=df['Survived'].count()
print("Total number of passengers =",a)

b=df['Survived'].value_counts()
print("Passergers survived(1)/ died(0) =",b)

c=(df['Survived']==1).sum()
print("Total Survived passengers =",c)

#Survival rate of passengers
s= c/a*100
print("Survival rate of passengers =",s)

#Analysis Answer=  The overall survival rate of passengers on the Titanic is 67.21%, stating that only 67.21% of the passengers survived the disaster
```

```python
#Q2 Class and Survival
#•	How does the survival rate vary across different passenger classes (Pclass)?

a=df['Survived'].count()
print("Total number of passengers =",a)

c=(df['Survived']==1).sum()
print("Total Survived passengers =",c)

b=df.groupby('Pclass')['Survived'].count()
print("Total Number of passengers in each class=",b)


#Class 1 passenger count
e=df['Pclass']
f=(e==1).sum()
print("Passengers in first class =",f)

#Survival % in class 1
g= f/c*100
print("SURVIVAL % IN CLASS 1 =",g)

#Class 2 passenger count
x=df['Pclass']
y=(x==2).sum()
print("Passengers in first class =",y)

#Survival % in class 2
z= y/c*100
print("SURVIVAL % IN CLASS 2 =",z)

#Class 3 passenger count
t=df['Pclass']
p=(t==3).sum()
print("Passengers in first class =",p)

#Survival % in class 3
m= p/c*100
print("SURVIVAL % IN CLASS 3 =",m)

#Analysis= Survival rate is highest in 1st class of passengers ie 128.8% and lowest in 8.13%.
```

```python
#4.	Family Size Impact
#•	How does the number of siblings/spouses (SibSp) and parents/children (Parch) affect the survival rate?

x=df['SibSp'].sum()
print("Total number of Sibsp = ",x)

y=df['Parch'].sum()
print("Total number of Parch = ",y)

c=(df['Survived']==1).sum()
print("Total Survived passengers =",c)

# SURVIVAL RATE OF (SibSp)
s= x/c*100
print("SURVAL % RATE OF SIBSP =",s)

# SURVIVAL RATE OF (Parch)
a= y/c*100
print("SURVAL % RATE OF Parch =",a)

#Analysis= The % of (SibSp) and survial rate is= 69.10% & % of (Parch) and survival rate is=70.73%. Stating that there are more than half % of the SibSp and Parch are outsid Titanic, waiting for their family member.
```

```python
#Q5 Fare Analysis:
#•Is there a relationship between fare and survival?

x=df['Fare'].sum()
print("Total Amount of fare paid by passengers as a whole =",x)

c=(df['Survived']==1).sum()
print("Total Survived passengers =",c)

a=(df['Survived']==0).sum()
print("Total NOT SURVIVED passengers =",a)

d= c/x*100
print("Relation b/w fair paid by the passengers who SURVIVED =",d)

e= a/x*100
print("Relation b/w fair paid by the passengers who NOT SURVIVED =",e)


#Analysis= Therefore we can state that 41.66% of passengers who paid the fair- NOT SURVIVED, and 85.42% of passengers SURVIVED, from the total amount of fair paid as a whole.
```
