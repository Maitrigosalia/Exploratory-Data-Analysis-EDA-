```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv("C:\\Users\\INDIA\\Documents\\Maitri CTAG Python\\csv\\Rain fall data.csv")

df.info()
x=df['station_name']
y=df['average_rainfall']
plt.plot(x,y,marker="o",color="green",label="rainfall")
plt.xlabel("state")
plt.ylabel("year")
plt.title("rainfall data")
plt.legend()
plt.show()
```
