```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("C:\\Users\\INDIA\\Documents\\Maitri CTAG Python\\csv\\Weekly restaurant order.csv")

df.info()
df.describe()

x1=df['food']
y1=df['exp']
x2=df['drink']
y2=df['expense']
plt.plot(x1,y1,marker= "x",color="green",linestyle="dotted",label="food and expense")
plt.plot(x2,y2,marker = "*",color="red",label="drink and expense")
plt.x1label("Food")
plt.x2label("Drink")
plt.y1label("Food and expense")
plt.y2label("Drink and expense")
plt.title("Weekely restaurant order")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv("C:\\Users\\INDIA\\Documents\\Maitri CTAG Python\\csv\\covid data.csv")

df=pd.DataFrame(df)

x=df['Month']
m=df['mask']
s=df['sanitizer']
h=df['handwash']
g=df['gloves in pair']
plt.plot(x,m,marker="o",linestyle="dotted",color="blue",label="mask")
plt.plot(x,s,marker="*",color="green",label="sanitizer")
plt.plot(x,h,marker="^",color="purple",label="handwash")
plt.plot(x,g,marker=".",color="red",label="gloves")
plt.xlabel("Month")
plt.ylabel("mask,sanitizer,handwash and gloves data")
plt.title("Covid data")
plt.legend()
plt.show()


```
