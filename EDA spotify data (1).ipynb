```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv("C:\\Users\\INDIA\\Documents\\Maitri CTAG Python\\csv\\Most Streamed Spotify Songs 2024.csv")
print(df)
df.head(10)
```

```python
df=df.drop(['Release Date','ISRC'],axis=1)
df.head(10)
```

```python
df=df.rename(columns={'All Time Rank':'WorldWide famous'})
df.head(5)
```

```python
df=pd.DataFrame(df)
df=df.head(5)
print(df)
x=df['Artist']
print(x)
y=df['Track Score']
plt.plot(x,y,marker="*",color="green",label="track records")
plt.xlabel("artist")
plt.ylabel("track score")
plt.title("data")
plt.show()
```

```python
df=df.head(10)

x=df['Artist']
y=df['Spotify Playlist Reach']
y1=df['YouTube Playlist Reach']
plt.plot(x,y,marker="*",color="blue",label="spotify reach")
plt.plot(x,y1,marker="o",color="red",label="youtube reach")
plt.xlabel("artist")
plt.ylabel("spotify vs yt")
plt.title("paltform stats")
plt.legend()
plt.show()
```

```python
#Plot a bar chart to compare the total Spotify streams of the top 10 songs.

df['Artist'].value_counts().nlargest(10).plot(kind="bar",figsize=(10,8),color="red")
plt.title("TOP ARTISTS OF 2024")
plt.ylabel("Streams")
plt.xlabel("Artist")
```

```python
#Create a scatter plot showing the relationship between release dates and Spotify popularity scores.


plt.scatter(df['Artist'],df['Spotify Streams'])
plt.show()
```

```python
#Use a grouped bar chart to compare streams (Spotify, YouTube, Pandora, Soundcloud) for the top 5 songs.

x=df['Artist']
y=df['Spotify Streams']
y1=df['YouTube Views']
y2=df['TikTok Views']
y3=df['Apple Music Playlist Count']
plt.plot(x,y,marker="o",color="blue",label="spotify streams")
plt.plot(x,y1,marker="*",color="green",label="yt views")
plt.plot(x,y2,marker="^",color="red",label="tiktok views")
plt.plot(x,y3,marker="x",color="purple",label="apple music")
plt.legend()
plt.show()
```

```python
#Generate a scatter plot to investigate the correlation between Spotify playlist reach and Spotify streams.

plt.plot(df['Spotify Streams'],df['Spotify Playlist Reach'],color="red",marker="*")
plt.title("streams vs reach")
plt.xlabel("streams")
plt.ylabel("reach")
plt.show()
```

```python
#Platform Engagement Comparison. Create a stacked bar chart showing engagement (likes, posts, streams) across platforms like Spotify, YouTube, and TikTok for the top 5 songs.

x=df['Artist']
y=df['YouTube Likes']
y1=df['TikTok Likes']
plt.plot(x,y,marker="*",color="green",label="youtube likes")
plt.plot(x,y1,marker="^",color="blue",label="tiktok likes")
plt.title("youtube vs tiktok likes")
plt.xlabel("artist")
plt.ylabel("comparison")
plt.legend()
plt.show()
```

```python
#Plot a histogram to analyze the distribution of track scores across all songs.

x=df['Artist']
y=df['Track Score']
plt.plot(x,y,marker="*",color="blue",label="Artist and track score")
plt.legend()
plt.show()
```

```python
#Plot a horizontal bar chart to visualize the Shazam counts of the top 10 tracks.

df=df.head(10)
print(df)

x=df['Artist']
y=df['Shazam Counts']
plt.barh(x,y,color="green",label="shazam count")
plt.title("Artist and counts")
plt.xlabel("counts")
plt.ylabel("artist")
plt.legend()
plt.show()
```

```python
#Create a scatter plot to assess whether AirPlay spins correlate with Spotify streams.
import matplotlib.pyplot as plt

plt.scatter(df['AirPlay Spins'],df['Spotify Streams'],marker="^",color="red",label="airplay spins and spotify streams")
plt.xlabel("airplay")
plt.ylabel("spotify")
plt.title("airplay vs spotify")
plt.legend()
plt.show()
            
```

```python
#Create a pie chart showing the share of streams by album name.
df=df.head(10)

l=df['Album Name']
print(l)
y=df['Spotify Streams']
plt.pie(y,labels=l)
plt.legend(title="Share of streams")
```

```python
#TikTok Metrics Analysis. Compare TikTok likes, views, and posts using a grouped bar chart for the top 5 songs with the highest TikTok engagement.

x=df['Artist']
y=df['TikTok Posts']
y1=df['TikTok Likes']
y2=df['TikTok Views']
plt.xlabel("Artist")
plt.ylabel("count")
plt.title("tiktok metrics")
plt.plot(x,y,marker="*",color="green",label="posts")
plt.plot(x,y1,marker="x",color="red",label="likes")
plt.plot(x,y2,marker="o",color="blue",label="views")
plt.legend()
plt.show()
```

```python
#Spotify Popularity vs. All-Time Rank

x=df['Spotify Popularity']
y=df['All Time Rank']
plt.plot(x,y,marker="*",color="red",label="spotify vs all time rank")
plt.xlabel("spotify popularity")
plt.ylabel("all time rank")
plt.title("comparision")
plt.legend()
plt.show()
```

```python
#Create a grouped bar chart comparing the playlist reach of Spotify, Deezer, and Amazon Music for the top 10 songs.

df=df.head(5)
#print(df)
x=df['Artist']
y=df['Spotify Streams']
y1=df['Deezer Playlist Reach']
y2=df['Pandora Streams']
plt.bar(x,y,color="red",label="spotify streams")
plt.bar(x,y1,color="green",label="deezer reach")
plt.bar(x,y2,color="blue",label="pandora streams")
plt.xlabel("artist")
plt.ylabel("comaprision")
plt.title("grouped chart")
plt.legend()
plt.show()
```

```python
#Plot a side-by-side bar chart to compare the airplay and SiriusXM spins for each track.

x=df['Artist']
y=df['AirPlay Spins']
plt.plot(x,y,marker="*",color="green",label="airplay")
plt.xlabel("artist")
plt.ylabel("airplay")
plt.legend()
plt.show()


x=df['Artist']
y=df['SiriusXM Spins']
plt.plot(x,y,marker="x",color="red",label="XM spins")
plt.xlabel("artist")
plt.ylabel("XM spins")
plt.legend()
plt.show()
```

```python
#Create a stacked bar chart to show the total engagement (streams, views, likes) for each platform (Spotify, YouTube, TikTok) across all songs.

x=df['Artist']
y=df['TikTok Posts']
y1=df['TikTok Likes']
y2=df['TikTok Views']
plt.bar(x,y,color="red",label="posts")
plt.bar(x,y1,color="blue",label="likes")
plt.bar(x,y2,color="green",label="views")
plt.xlabel("artist")
plt.ylabel("metrics of tiktok")
plt.legend()
plt.show()
```

```python
#Use a grouped bar chart to visualize each artist’s total engagement across platforms like Spotify, YouTube, and TikTok.

x=df['Artist']
y=df['Spotify Streams']
y1=df['YouTube Views']
y2=df['TikTok Views']
plt.bar(x,y,color="red",label="spotify")
plt.bar(x,y1,color="green",label="youtube")
plt.bar(x,y2,color="blue",label="tiktok")
plt.xlabel("artist")
plt.ylabel("comparision of artist amoung platforms")
plt.title("artist engagement on diff paltforms")
plt.legend()
plt.show()
```
